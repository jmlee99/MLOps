# **Section 2. Part2. 머신러닝**

## **1. Maching Learning**

- 머신 러닝은 지속적으로 경험을 바탕으로 주어진 작업을 개선해 나아가는 머신이다.
- 머신 러닝안에는 **지도학습**, **비지도 학습**, **강화 학습**이 있다.

**1-1. 지도 학습**<br/>
> 지도 학습은, 학습 알고리즘이 <span style='background-color:#fff5b1'>**정답 값이 있는 입력과 출력 데이터**</span>를 사용해서 알고리즘을 학습하는 것이다.<br/>
> **Ex)** 개와 고양이를 분류할 때, 개와 고양이라는 라벨(정답)을 달고 학습하는 것이다.<br/>
> 성능은 모델이 예측한 결과와 실제 결과를 비교해서 평가<br/>
<br/>
> **1-1-1. 지도학습의 두가지 기술**<br/>
>> **(1) 분류**<br/>
>> 두 가지의 결과가 있을 때 발생한다. <br/>
>> **Ex)** 암에 걸린 환자인지 아닌지, 고객이 행복한지 안한지<br/>
>> 즉, 두 개 이상의 결과가 있을 때를 말한다. **{불연속의 출력(범주형)}**<br/>
>>
<br/>
>> **(2) 회귀**<br/>
>> 이때는 연속되는 결과를 예측한다.<br/>
>> **Ex)** 주가, 소매 주간 매출액 **{연속의 출력(수치형)}**

**1-2. 비지도 학습**<br/>
> 라벨(정답)이 없는 데이터를 통해 알고리즘을 학습하는 것이다.<br/>
> Ex) 단순 이미지를 숨겨진 패턴을 스스로 학습하도록 요청하는 것<br/>
<br/>
> **1-2-1. 비지도 학습, 군집화**<br/>
>> **(1) 군집화**<br/>
>> 주로 군집화(그룹화)를 위해 사용<br/>
>> **Ex)** 서로 다른 무리로 고객들을 그룹 지어야 할때 $=>$ 특정 그룹을 위해 이벤트를 진행할 때<br/>
>>

**1-3. 강화 학습**<br/>
> 알고리즘이 누적된 보상을 극대화하기 위해 조치를 취하는 것이다.<br/>
> 보상과 패널티를 통해 시행착오를 배우려고 한다고 생각해보자.<br/>
<br/>
>> **Ex)** <br/>
>> 에이전트(AI)가 작동하거나 환경에 배치되어 그 환경에서 에이전트가 조치를 취하는 것이다.<br/>
>> **에이전트는 시간이 지남에 따라 많은 시행착오를 통해 그 환경을 개선 시킬 것이다.**<br/>
>> <br/><br/>
<span style='background-color:#fff5b1'>**강화 학습은 인간의 경험과 같다**</span> <br/>

## **4. AWS의 지역과 가용성 영역**

### **4-1. 가용성 영역과 지역의 차이점**

1. **지역**이란?
- AWS 지역은 여러 가용성 영역이나 데이터 센터를 포함하고 있는 지리적 위치
- 이 지역 안에 **가용성 영역**들이 존재한다.
    * 즉, 배포하고 싶은 지역을 확인하는 것이 중요하다.
    * US, 아프리카에 배포를 한다면, 모든 서비스는 고객과 가까운 서버에서 실행해야한다.
    * 아시아에 서버를 두면, 고객과 서비스의 상호 작용 관계에서 지연 시간을 경험할 수 있다.
    
2. **가용성 영역**이란
- 간단히 말해, 조정된 데이터 센터이다.<br/>
<br/>
- 파란색이 지역, A,B,C가 세 개의 가용성 영역이다.
- 하나의 지역안에서 A를 사용하다가 전력이 다하면, B나 C를 여분으로 사용하다가 A가 복구되면 돌아올 수 있다.
- 즉, AWS가 자사의 서비스를 매우 높은 내구성과 높은 가용성으로 시장에 출시한다.

# **Section 2. Part7. AMAZON S3**

## **1. AMAZON S3**
- 스토리지 서비스의 줄임말로, **기업과 개인이 데이터를 간단하게 저장하고 보호**할 수 있는 스토리지 서비스이다.
> - 실제로 물리적인 서버나 저장 매체를 살 필요가 없다.<br/>
> - S3에 데이터를 직접적으로 업로드 할 수 있다.<br/>
> - S3와 결합된 아마존 세이지 메이커 사용 및 머신러닝 학습을 위해, SageMaker는 데이터를 S3에 업로드 하도록 요구한다.
>> **강력한 장점**<br/>
>> 기업은 데이터를 정리하고 세밀한 접근 통제를 구성한다.(특정 인원으로 제한)

## **2. 버킷**
- 버킷은 아마존 S3에 저장된 객체를 담는 컨테이너이고, 모든 객체는 버킷에 담겨있다.
- 각각의 버킷은 전체적 고유 이름을 갖게 된다.<br/>즉, **다른 사람과 공유하는 같은 이름의 버킷을 가질 수 없다.**
- 버킷에 무한대의 데이터를 저장할 수 있고, 버킷 내의 각 객체는 5TB의 데이터 용량으로 제한된다.<br/>**사실상 버킷 안에 무제한의 스토리지가 있는 셈이다.**
>**Ex)**<br/>
> 만약 images/mycat.jpg(객체) $>>$(저장)$>>$ sagemakerpractical(버킷)<br/>
> http://{(버킷)}.s3.amazonaws.com/{(객체)} 이 URL로 접속하면 내 버킷에 접근할 수 있다.<br/>여기서는 http://sagemakerpractical.s3.amazonaws.com/images/mycat.jpg 가 된다.

- 이러한 URL을 가지게 되면 실제로 데이터를 검색하고, 모델을 학습할 수 있다. 또한, 테스팅 데이터 또한 업로드 할 수 있다.

## **3. S3 스토리지 클래스**
- 데이터 저장 관련 비용이 발생할 때, 목표로 하는 스토리지 클래스 선택에 신중해야한다.<br/>
|||
|---|---|
|S3 Standard|- 범용적으로 많이 사용<br/> - S3의 데이터에 자주 접속한다면 S3 스탠다드 유형을 선택|
|S3 Intelligent-Tiering|- 다양한 액세스 패턴을 가진 데이터와 잘 작동<br/>- 일주일 안에 자주 접속하고, 그 이후로 데이터를 한 달 동안 액세스 없이 방치하는 경우|
|S3 Standard-Infrequent Access|- 단순히 오랜 기간 보관하지만 데이터 액세스하는 빈도가 낮은 경우에 사용|
|S3 One Zone-Infrequent Access|- 단순히 오랜 기간 보관하지만 데이터 액세스하는 빈도가 낮은 경우에 사용<br/> - 내 데이터가 한 개의 영역에서만 사용 가능|
|Amazon S3 Glacier|- 매우 장기적인 아카이브 데이터 경우<br/> 예를 들어, 고객의 의료 정보 같은 것<br/>바로 엑세스 할 계획이 없는 경우|

- 처음에는 Genral 한달 뒤쯤 Infrequent Access 몇 달 뒤 Glacier로 보관하는 것도 가능하다.
- 강의에서는 S3 스탠다드를 주로 사용

- (나중을 위해)각 스토리지 클래스는 이러한 특징을 가진다

## **2. 계정 접근 관리 (IAM)**
- 사용자가 안전하게 AWS서비스에 액세스하고 관리할 수 있도록 한다.
- IAM은 AWS사용자 및 그룹을 만들고 누가 EC2, S3 버킷 등에 액세스 할 수 있는지 등의 액세스 권한을 지정하는 것이다.
- 항상 사용하는 것은 매우 위험하다.
> 자격에 대한 증명(KEY)를 도난당한 경우 다른 사람이 사용해 비싼 청구서를 받을 수도 있다.<br/>
> **따라서 AWS는 일상적인 작업에 루트 사용자를 사용하지 말고 대신, IAM 역할 또는 계정 접근 관리 역할을 생성한 다음 루트 사용자의 자격 증명을 잠글 것을 권장한다.**<br/>
