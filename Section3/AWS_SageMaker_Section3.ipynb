## **1. MAE(평균 절대 오차)**
1. **모델예측과 실제 값 사이의 차이의 절대값을 걔산한다.**
2. **모든 오차 절대값의 합을 구한다.**
3. **샘플의 총 개수(n)으로 나눈다.**



4. **MAE가 0에 가까울수록 모델이 완벽하다는 것이다.**

## **2. MSE(평균 제곱 오차)**
- **MAE와 매우 유사하다.** 

1. **일반적으로 MAE보다 수치가 크다 (제곱을 하기 때문에)**
2. **오차가 클수록 MSE의 값은 훨씬 커진다. == 이상치가 (편향된 정도?)**



## **3. RMSE(평균 제곱근 오차)**
- **MSE에 비해 쉽게 해석가능**
    * 제곱근을 구해 MAE와 비교적 비슷한 수치로 돌아간다.
        * 평균으로부터의 분산이다.
    * 예를 들어, 직원의 급여를 예측
    > **평균 제곱근 오차의 단위가 실제 급여 단위라고 예상 가능**



## **4.MAPE(평균 절대비 오차)**
- **MAE를 %로 확인하기 좋다.**
- **한계가 명확**
    * 만약 데이터 포인트 값이 0이면 계싼이 엉망이된다.

# **3. $R^2$ or 결정 계수**



- **결정계수는 X의 독립 변수에 의해 설명되는 출력 변수 y의 분산 비율**
    * ***EX)*** 보험 비용의 얼마나 많은 분산이 연령의 분산으로 설명되는가?
    * $R^2$이 80과 같다면 보험 비용 증가의 80%가 신청자 연령의 증가에<br>
      원인을 둔다고 생각하면 된다. 
- **$R^2$이 높다면 데이터의 분산을 잘 파악해서 모델이 만들어졌다고 생각**
- **$R^2$이 낮다면 데이터의 분산을 잘 파악하지 못한 것**
    * 이런 경우 모델의 독립 변수를 추가하거나 늘리는 등 조치를 취해야함
    * ***EX)*** 연령(x)이 아무리 바뀌어도 보험료(y)가 달라지지 않으면,<br/>
      $R^2$이 낮다는 것 즉, 연령이 보험료 예측에 쓸모가 없다는 걸 예측
      
## **$R^2$의 단점**
- **단순 $R^2$은 쓸모없는 독립변수라도 늘리면 $R^2$값은 상승한다.**
    * 보험료 예측에 전혀 필요없는 자동차 색이 들어가도 $R^2$ 값은 올라간다.
    * 우리는 조정된 $R^2$이 필요하다 (업그레드 버전)
    

- 이 식은 쓸모없는 변수를 추가하면 패널티를 받아 $R^2$값이 오히려 떨어진다.

# **2. L1 정규화(Lasso), L2 정규화(Ridge)**

- **과적합을 피하고 벗어나려 하는 것이다.<br/>==> 과적합을 피하기 위해 오류를 입력하는 것이다.**



- **이 모델은 훈련 데이터와 시험 데이터 사이에 실질적으로 높은 분산, 또는 높은 변량이 존재하는 것이다.**
- **이 모델은 Train 데이터에 과적합 되어 일반화에 실패를 한 것이다.**



- **이러한 모델에 적정한 선을 찾기 위해 모델의 경사를 바꾸는 것이다.**
- **즉, 분산을 향상하기 위해 편향을 높이려고 선의 경사를 바꾸는 것이다.**

**경사를 줄이는 공식**





- **위와 같이 alpha = 값을 변형 시켜 기울기를 조절할 수 있다.**

- 만약 쓸모없는 변수가 있다면, 예측에 많은 기여를 하지 않는 독립변수가 있다.<br/> == **Lasso**

**차이점**
- **L1(Lasso)는 몇몇 특성을 0으로 가게한다.**
- **L2(Ridge)는 모든 특성은 동일하지만 가중치를 더한다., 어떤 특성도 0으로 만들지 X**

- L1 : 중요하지 않은 특성들이 있다고 생각들어 과적합 방지를 위해 특정 값을 없애기 위해 사용
- L2 : 중요하지 않은 특성들이 없어 모두 들고 가고 싶지만 과적합 방지할때 사용
